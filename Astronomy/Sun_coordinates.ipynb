# coordinates of the Sun

```python
import numpy as np
import matplotlib.pyplot as plt
import smplotlib
```

$\alpha_\odot=360^\circ\frac{t}{365.25\mathrm{\,d}}$

$\delta_\odot=\arctan\left(\sin\alpha_\odot\cdot\tan{i}\right)$

```python
day = np.linspace(0,365.24,1000)
alpha_sun = 360*day/365.25

delta = np.arctan(np.sin(alpha_sun*np.pi/180)*np.tan(23.5*np.pi/180))*180/np.pi
```

```python
fig = plt.figure(dpi=150,figsize=(16,6))

ax1 = fig.add_subplot(121)
ax1.plot(day,alpha_sun)
ax1.set_xticks(np.arange(0, 365.25, 30))
ax1.set_xlabel('Starting from the vernal equinox (day)') 
ax1.set_ylabel(r'$\alpha_\odot$ (deg)')
ax1.grid(True)

ax2 = fig.add_subplot(122)
ax2.plot(day,delta)
ax2.set_xticks(np.arange(0, 365.25, 30))
ax2.set_xlabel('Starting from the vernal equinox (day)') 
ax2.set_ylabel(r'$\delta_\odot$ (deg)')
ax2.grid(True)
```
